```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### create calendar df

```python
calend_df = pd.DataFrame({'date':pd.date_range(start='1/1/1992', end='12/31/2015')})
calend_df['year'] = calend_df['date'].dt.year
calend_df['month'] = calend_df['date'].dt.month
calend_df['weekday'] = calend_df['date'].dt.weekday #Monday=0 

calend_df
```

### import weather data

```python
file_weather = 'cloud-data/digitalrnd-projects-ireland/Vaccine/Research/Magellan/users/XLO/test/temp/weather_data_forest.csv.gz'
df_weather = pd.read_csv(file_weather, compression='gzip', sep=',', header=0, index_col=0)
df_weather['date'] = pd.to_datetime(df_weather['date'], format='%Y-%m-%d')

df_weather
```

### import lightnings data

```python
file_lightnings = 'cloud-data/digitalrnd-projects-ireland/Vaccine/Research/Magellan/users/XLO/test/light/light_data_forest.csv.gz'
df_lightnings = pd.read_csv(file_lightnings, compression='gzip', sep=',', header=0)
df_lightnings['#ZDAY'] = pd.to_datetime(df_lightnings['#ZDAY'], format='%Y-%m-%d')
df_lightnings = df_lightnings.rename(columns={'#ZDAY':'date', 'TOTAL_COUNT':'#lightnings'})
df_lightnings['lightnings'] = 1

df_lightnings
```

### import wildfires data

```python
file_wildfires = 'cloud-data/digitalrnd-projects-ireland/Vaccine/Research/Magellan/users/XLO/test/wildfires/wildfires_data_forest.csv.gz'
df_wildfires = pd.read_csv(file_wildfires, compression='gzip', sep=',', header=0)

df_wildfires = df_wildfires.groupby(by=['PROCLAIMED', 'DISCO_DATE']) \
                                .agg({'FOD_ID':'count'}).rename(columns={'FOD_ID':'#wildfires'}).reset_index()
df_wildfires['DISCO_DATE'] = pd.to_datetime(df_wildfires['DISCO_DATE'], format='%Y-%m-%d')
df_wildfires = df_wildfires.rename(columns={'DISCO_DATE':'date'})

df_wildfires
```

### import forests with states

```python
file_forests = 'cloud-data/digitalrnd-projects-ireland/Vaccine/Research/Magellan/users/XLO/test/wildfires/us_forests_ext_d100+state.csv.gz'
df_forests = pd.read_csv(file_forests, compression='gzip', sep=',', header=0, index_col=0)

selection = [295373010328, 295469010328, 295371010328, 295471010328]

df_forests.loc[df_forests['PROCLAIMED'].isin(selection)]
```

### create Coconino National Forest df

```python
df_coco = pd.merge(left=calend_df, right=df_weather.loc[df_weather['PROCLAIMED']==295373010328, ['date', 'TMAX_mean', 'TMIN_mean', 'TMAX_mean_10D', 'TMAX_mean_30D', 'TMIN_mean_10D', 'TMIN_mean_30D','PRCP_mean', 'PRCP_mean_10D', 'PRCP_mean_30D']], on='date', how='left')
df_coco = pd.merge(left=df_coco, right=df_lightnings.loc[df_lightnings['PROCLAIMED']==295373010328, ['date', 'lightnings']], on='date', how='left')
df_coco = pd.merge(left=df_coco, right=df_wildfires.loc[df_wildfires['PROCLAIMED']==295373010328, ['date', '#wildfires']], on='date', how='left')
df_coco
```

```python
df_coco.isnull().sum()
```

```python
df_coco = df_coco.fillna({'lightnings':0, '#wildfires':0})
df_coco['#wildfires_1D'] = df_coco['#wildfires'].rolling(window=1, closed='left').sum()
df_coco['#wildfires_10D'] = df_coco['#wildfires'].rolling(window=10, closed='left').sum()
df_coco['#wildfires_30D'] = df_coco['#wildfires'].rolling(window=30, closed='left').sum()
df_coco = df_coco.fillna({'#wildfires_1D':0, '#wildfires_10D':0, '#wildfires_30D':0})
df_coco['target'] = df_coco['#wildfires'].apply(lambda x: 0 if x==0 else 1)
df_coco
```
