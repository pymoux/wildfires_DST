```python
import pandas as pd
```

from https://www.ncei.noaa.gov/access/metadata/landing-page/bin/iso?id=gov.noaa.ncdc:C00773

```python
files = [f'/content/drive/MyDrive/DST_Wildfires/lightening_data/data/nldn-tiles-{i}.csv.gz' for i in range(1992, 2016)]
```

```python
df_list = []
for f in files:
    print(f[-11:])
    df = pd.read_csv(f,compression='gzip', sep=',', header=2)
    # index_col=False
    df_list.append(df)

df_out = pd.concat(df_list, axis=0, ignore_index=True)
df_out.shape
```

```python
out_path = "/content/drive/MyDrive/DST_Wildfires/lightening_data/lightnings.csv.gz"
df_out.to_csv(out_path, compression='gzip')
print('file ok')
```
