```python
import pandas as pd
```

```python
df_state = pd.read_csv('/content/drive/MyDrive/DST_Wildfires/counties/FIPS.txt',
                       sep='\t',
                       header=None,
                       names=['State_FIPS_code', 'State_name', 'State_abbr'],
                       skiprows=15,
                       skipfooter=3199,
                       engine='python',
                       )

df_state['State_name'] = df_state['State_name'].str.title()
df_state['State_FIPS_code'] = df_state['State_FIPS_code'].apply(lambda x: '{:02d}'.format(x))

df_state.head()
```

```python
df_county = pd.read_csv('/content/drive/MyDrive/DST_Wildfires/counties/FIPS.txt',
                       sep='\t',
                       header=None,
                       names=['County_FIPS_code', 'County_name', 'County_comment'],
                       skiprows=78,
                       )

#df_state['State'] = df_state['State'].str.title()
df_county['County_FIPS_code'] = df_county['County_FIPS_code'].apply(lambda x: '{:05d}'.format(x))

df_county.head(15)
```

```python
df_county['State_code'] = df_county['County_FIPS_code'].str[0:2]
#df_county.sample(10)
df = pd.merge(df_county, df_state, left_on='State_code', right_on='State_FIPS_code')
df = df.drop('State_code', axis=1)
df.sample(15)
```
